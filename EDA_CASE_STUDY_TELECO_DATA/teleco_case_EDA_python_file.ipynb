***feature Engineering***

Categoring Numerical Columns to do Univariate analysis with churn

```python
# allotting 1 to yes and 0 to no for the Churn Column to do Further Analysis

df['Churn_flagging'] = df['Churn'].map({"Yes": 1, 'No':0})
df.head(1)
```

```python
# tenure_flagging 
df['tenure_flagging'] = np.where((df['tenure'] > 0) & (df['tenure'] < 12) , 'New',
        np.where((df['tenure'] > 13) & (df['tenure'] < 24), 'Mid',
        np.where((df['tenure'] > 24) & (df['tenure'] < 48), 'Loyal', 
                 'Very Loyal'
                )))
```

```python
df.head(1)
```

```python
# Monthly Charges flagging
df['montly_charges_flagging'] = np.where(df['MonthlyCharges'] <= 35, 'Low',
        np.where((df['MonthlyCharges'] > 35) & ( df['MonthlyCharges'] <= 70), 'Mid',
        'High'))
```

```python
df.head(1)
```

```python
# Total Charges Flagging

df['total_amt_flag'] = pd.cut(df['TotalCharges'], bins=[0, 1000, 3000, 5000, 9000], labels=['Low', 'Mid', 'High', 'very High'])
```

```python
df.head(1)
```

```python
df[['Churn', 'Churn_flagging', 'tenure' ,'tenure_flagging', 'MonthlyCharges' ,'montly_charges_flagging' , 'TotalCharges' ,'total_amt_flag']].head(1)
```

***Multivariate Analysis***

```python
# Contract + Tenure + Churn

pivot = df.pivot_table(
    values='Churn_flagging',
    index='Contract',
    columns='tenure_flagging',
    aggfunc='mean'
) * 100

sns.heatmap(pivot, annot=True, fmt=".1f", cmap="coolwarm")

plt.title("Churn % by Contract & Tenure")
plt.show()
```

***“Customers with month-to-month contracts and low tenure have the highest churn rates, while long-term contract customers with higher tenure show the lowest churn.”***

```python
pivot2 = df.pivot_table(
    values='Churn_flagging',
    index='InternetService',
    columns='montly_charges_flagging',
    aggfunc='mean'
) * 100

sns.heatmap(pivot2, annot=True, fmt=".1f", cmap="YlOrRd")
plt.title("Churn % by Charges & Internet Service")
plt.show()
```

***“Customers using fiber internet services with high monthly charges exhibit the highest churn rates, indicating dissatisfaction among high-paying users.”***

***“High-paying, low-tenure, month-to-month customers are the most likely to churn.”***
