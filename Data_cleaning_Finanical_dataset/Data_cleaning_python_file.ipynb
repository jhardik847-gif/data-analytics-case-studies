Till now We Focus on Identify Problems in this messy Dataset.

```python
# Dropping Duplicated Values

messy_data.drop_duplicates(inplace=True)
```

```python
# Checking duplicates values again

messy_data[messy_data.duplicated()]
```

```python
# dropping rows where customer id is null

messy_data.dropna(subset=['Customer_ID'], inplace=True)
```

```python
messy_data.isnull().sum()/len(messy_data)*100
```

```python
# generating transacation id for null values

# Total rows count karein
n = len(messy_data)

# Nayi IDs generate karein (T00001 format mein)
messy_data['Transaction_ID'] = [f'T{str(i+1).zfill(5)}' for i in range(n)]
```

```python
messy_data.isnull().sum()/len(messy_data)*100
```

```python
# changing price data types and working on null values and '$'

messy_data['Price'].str.startswith('$').value_counts()
```

```python
messy_data["Price"] = messy_data['Price'].str.replace('$', '', regex=False)
```

```python
messy_data['Price'].sort_values().unique()
```

```python
messy_data['Price'] = messy_data['Price'].astype(float)
```

```python
messy_data.info()
```

```python
messy_data['Price'].sort_values().unique()
```

```python
messy_data['Price'] = messy_data['Price'].fillna(messy_data['Price'].median())
```

```python
messy_data.info()
```

```python
# working in Quantity Columns
```

```python
messy_data['Quantity'].sort_values().unique()
```

```python
messy_data['Quantity'] = messy_data['Quantity'].fillna(messy_data['Quantity'].median())
```

```python
messy_data['Quantity'].sort_values().unique()
```

```python
messy_data['Quantity'] = messy_data['Quantity'].astype(int)
```

```python
messy_data['Quantity'] = messy_data['Quantity'].apply(lambda x: x if x >= 0 else 0 )
```

```python
messy_data['Quantity'].sort_values().unique()
```

```python
messy_data['Transaction_Status'] = messy_data['Transaction_Status'].fillna('completed')
```

```python
messy_data['Transaction_Status'].unique()
```

```python
messy_data['Transaction_Status'] = messy_data['Transaction_Status'].replace({'completed': 'Completed',
                                                                            'complete': 'Completed'})
```

```python
messy_data['Transaction_Status'].unique()
```

```python
messy_data['Payment_Method'].unique()
```

```python
messy_data['Payment_Method'] = messy_data['Payment_Method'].replace({'pay pal': 'Paypal', 'PayPal ': 'Paypal',
                                                                    'creditcard': 'CreditCard', 'credit card': 'CreditCard' })
```

```python
messy_data['Payment_Method'] = messy_data['Payment_Method'].replace({'Paypal': 'PayPal'})
```

```python
messy_data['Payment_Method'] = messy_data['Payment_Method'].replace({'Credit Card': 'CreditCard'})
```

```python
messy_data['Payment_Method'].unique()
```

```python
messy_data['Product_Name'].unique()
```

```python
messy_data['Product_Name'] = messy_data['Product_Name'].str.lower()
```

```python
messy_data['Product_Name'] = messy_data['Product_Name'].apply(lambda x: 'coffee' if x.startswith('c') else
                                                             'headphones' if x.startswith('h') else 
                                                              'laptop' if x.startswith('l') else
                                                              'smartphone' if x.startswith('s') else
                                                              'other')
```

```python
messy_data['Product_Name'].unique()
```

```python
messy_data['Transaction_Date'] = pd.to_datetime(messy_data['Transaction_Date'], errors='coerce')
```
